```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("/content/Lena_Image.png")  # Replace with your image path
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

# Compute grayscale histogram (M1)
hist_gray = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Normalize the histogram (M2: Probability distribution)
hist_gray_norm = hist_gray / hist_gray.sum()

# Compute histograms for color image (M1)
colors = ('b', 'g', 'r')  # OpenCV uses BGR order
hist_color = {c: cv2.calcHist([image], [i], None, [256], [0, 256]) for i, c in enumerate(colors)}

# Normalize color histograms (M2)
hist_color_norm = {c: hist / hist.sum() for c, hist in hist_color.items()}

# Plot histograms (M1 & M2)
plt.figure(figsize=(12, 6))

# 🔹 M1: Standard Histogram (Pixel Count)
plt.subplot(2, 2, 1)
plt.plot(hist_gray, color='black')
plt.title("M1: Grayscale Histogram")
plt.xlabel("Gray Level")
plt.ylabel("Number of Pixels")

plt.subplot(2, 2, 2)
for c, hist in hist_color.items():
    plt.plot(hist, color=c, label=f"{c.upper()} channel")
plt.title("M1: Color Histogram")
plt.xlabel("Gray Level")
plt.ylabel("Number of Pixels")
plt.legend()

# 🔹 M2: Normalized Histogram (Probability)
plt.subplot(2, 2, 3)
plt.plot(hist_gray_norm, color='black')
plt.title("M2: Normalized Grayscale Histogram")
plt.xlabel("Gray Level")
plt.ylabel("Probability of Occurrence")

plt.subplot(2, 2, 4)
for c, hist in hist_color_norm.items():
    plt.plot(hist, color=c, label=f"{c.upper()} channel")
plt.title("M2: Normalized Color Histogram")
plt.xlabel("Gray Level")
plt.ylabel("Probability of Occurrence")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original image and convert to grayscale
image = cv2.imread("/content/Lena_Image.png")  # Replace with your image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Modify brightness
dark_image = np.clip(gray_image - 50, 0, 255).astype(np.uint8)  # Darker
bright_image = np.clip(gray_image + 50, 0, 255).astype(np.uint8)  # Brighter

# Compute histograms
hist_orig = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
hist_dark = cv2.calcHist([dark_image], [0], None, [256], [0, 256])
hist_bright = cv2.calcHist([bright_image], [0], None, [256], [0, 256])

# Plot images
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(dark_image, cmap='gray')
plt.title("Dark Image")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(bright_image, cmap='gray')
plt.title("Bright Image")
plt.axis("off")

# Plot histograms
plt.subplot(2, 3, 4)
plt.plot(hist_orig, color='black')
plt.title("Original Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Pixel Count")

plt.subplot(2, 3, 5)
plt.plot(hist_dark, color='black')
plt.title("Dark Image Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Pixel Count")

plt.subplot(2, 3, 6)
plt.plot(hist_bright, color='black')
plt.title("Bright Image Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Pixel Count")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a grayscale image
image = cv2.imread("/content/Lena_Image.png", cv2.IMREAD_GRAYSCALE)  # Replace with your image

# Compute the 2D Fourier Transform using FFT
dft = np.fft.fft2(image)
dft_shift = np.fft.fftshift(dft)  # Shift zero frequency components to the center

# Compute magnitude spectrum
magnitude_spectrum = 20 * np.log(np.abs(dft_shift) + 1)  # Avoid log(0) by adding 1

# Compute the Inverse Fourier Transform
idft_shift = np.fft.ifftshift(dft_shift)  # Shift back
image_reconstructed = np.fft.ifft2(idft_shift)
image_reconstructed = np.abs(image_reconstructed)  # Take absolute values

# Plot the original image, magnitude spectrum, and reconstructed image
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title("Magnitude Spectrum")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(image_reconstructed, cmap='gray')
plt.title("Reconstructed Image (IFFT)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread("/content/Lena_Image.png", cv2.IMREAD_GRAYSCALE)  # Replace with your image

# Compute the 2D FFT and Magnitude Spectrum of Original Image
dft = np.fft.fft2(image)
dft_shift = np.fft.fftshift(dft)  # Shift the zero frequency component to the center
magnitude_spectrum = 20 * np.log(np.abs(dft_shift) + 1)  # Compute magnitude spectrum

# Rotate the image by 45 degrees
angle = 45
(h, w) = image.shape
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)  # Get rotation matrix
rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))  # Apply rotation

# Compute the 2D FFT and Magnitude Spectrum of Rotated Image
dft_rotated = np.fft.fft2(rotated_image)
dft_shift_rotated = np.fft.fftshift(dft_rotated)
magnitude_spectrum_rotated = 20 * np.log(np.abs(dft_shift_rotated) + 1)

# Plot images and magnitude spectra
plt.figure(figsize=(12, 6))

# Original image and its spectrum
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title("Magnitude Spectrum of Original Image")
plt.axis("off")

# Rotated image and its spectrum
plt.subplot(2, 2, 3)
plt.imshow(rotated_image, cmap='gray')
plt.title(f"Rotated Image ({angle}°)")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(magnitude_spectrum_rotated, cmap='gray')
plt.title(f"Magnitude Spectrum of Rotated Image ({angle}°)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
